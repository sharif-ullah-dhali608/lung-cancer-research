```python
# --- PHASE 1: INSTALLATION & IMPORTS ---
print("üì• Phase 1: Dependencies setup starting...")
!pip install optuna-integration[tfkeras] --quiet

import os, shutil, numpy as np, tensorflow as tf, optuna, time
import seaborn as sns, matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, auc
from tensorflow.keras import layers, Model
from tensorflow.keras.optimizers import Adam, Adamax, RMSprop 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from optuna.integration.tfkeras import TFKerasPruningCallback

print("‚úÖ Phase 1 Complete: Environment ready.")

# --- PHASE 2: GLOBAL CONFIGURATION ---
KAGGLE_INPUT_PATH = '/kaggle/input/lung-cancer-dataset-with-augmentation'
DATASET_ROOT = os.path.join(KAGGLE_INPUT_PATH, 'Lung Cancer Dataset')
TRAIN_DIR = os.path.join(DATASET_ROOT, 'Train')
ORIGINAL_TEST_DIR = os.path.join(DATASET_ROOT, 'Test') 
WORKING_DIR = '/kaggle/working'
VAL_DIR = os.path.join(WORKING_DIR, 'Val_50_percent')
FINAL_TEST_DIR = os.path.join(WORKING_DIR, 'Final_Test_50_percent')

IMAGE_SIZE = (224, 224); NUM_CLASSES = 3; SEED = 42
EPOCHS_PER_TRIAL = 8; INITIAL_EPOCHS = 25; FINE_TUNE_EPOCHS = 15 
N_TRIALS = 15; N_TRIALS_FREEZE = 8 

# YOLOv11 SEARCH SPACES (Based on Architecture Diagram)
C3K2_FILTERS_SEARCH = [64, 128, 256]
BACKBONE_DEPTH_SEARCH = [3, 4, 5] 
FREEZE_PERCENT_SEARCH = [10, 20, 30, 40, 50, 70, 80]

np.random.seed(SEED); tf.random.set_seed(SEED)
CLASS_NAMES = ['Bengin cases', 'Malignant cases', 'Normal cases']
print(f"‚úÖ Phase 2 Complete: Configured for {NUM_CLASSES} classes.")

# --- PHASE 3: YOLOv11 ARCHITECTURAL COMPONENTS (Exact as per Image) ---

def conv_bn_gelu(x, filters, kernel_size=3, strides=1):
    x = layers.Conv2D(filters, kernel_size, strides=strides, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    return layers.Activation('gelu')(x)

def c3k2_block(x, filters, shortcut=True):
    c = filters // 2
    x1 = conv_bn_gelu(x, c, kernel_size=1)
    x2 = conv_bn_gelu(x, c, kernel_size=1)
    y = conv_bn_gelu(x2, c, kernel_size=3)
    y = conv_bn_gelu(y, c, kernel_size=3)
    if shortcut: y = layers.Add()([x2, y])
    return conv_bn_gelu(layers.Concatenate()([x1, y]), filters, kernel_size=1)

def sppf_block(x, filters, k=5):
    c = filters // 2
    x = conv_bn_gelu(x, c, kernel_size=1)
    p1 = layers.MaxPooling2D(pool_size=k, strides=1, padding='same')(x)
    p2 = layers.MaxPooling2D(pool_size=k, strides=1, padding='same')(p1)
    p3 = layers.MaxPooling2D(pool_size=k, strides=1, padding='same')(p2)
    return conv_bn_gelu(layers.Concatenate()([x, p1, p2, p3]), filters, kernel_size=1)

def c2psa_block(x, filters):
    c = filters // 2
    x1 = conv_bn_gelu(x, c, kernel_size=1)
    x2 = conv_bn_gelu(x, c, kernel_size=1)
    attn = layers.GlobalAveragePooling2D()(x2)
    attn = layers.Reshape((1, 1, c))(attn)
    attn = layers.Dense(c, activation='sigmoid')(attn)
    x2 = layers.Multiply()([x2, attn])
    return conv_bn_gelu(layers.Concatenate()([x1, x2]), filters, kernel_size=1)

def build_yolov11_classifier(input_shape, num_classes, base_filters, stages, dropout_rate):
    inputs = layers.Input(shape=input_shape)
    x = conv_bn_gelu(inputs, base_filters // 2, strides=2)
    for i in range(stages):
        f = base_filters * (2**i) if i < 3 else base_filters * 4
        x = conv_bn_gelu(x, f, strides=2)
        x = c3k2_block(x, f)
    x = sppf_block(x, x.shape[-1])
    x = c2psa_block(x, x.shape[-1])
    x = layers.GlobalAveragePooling2D()(x)
    y = layers.Dense(base_filters * 2, activation='gelu')(x)
    y = layers.Dropout(dropout_rate)(y)
    return Model(inputs, layers.Dense(num_classes, activation="softmax")(y))

print("‚úÖ Phase 3 Complete: YOLOv11 Architecture defined.")

# --- PHASE 4: DATA LOADERS & SPLITTING ---
def get_loaders(bs):
    gen = ImageDataGenerator() 
    t = gen.flow_from_directory(TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=bs, class_mode='categorical', seed=SEED)
    v = gen.flow_from_directory(VAL_DIR, target_size=IMAGE_SIZE, batch_size=bs, class_mode='categorical', seed=SEED, shuffle=False)
    ts = gen.flow_from_directory(FINAL_TEST_DIR, target_size=IMAGE_SIZE, batch_size=bs, class_mode='categorical', seed=SEED, shuffle=False)
    return t, v, ts

print("üìÇ Splitting Test Data into Val (50%) and Final Test (50%)...")
for d in [VAL_DIR, FINAL_TEST_DIR]:
    if os.path.exists(d): shutil.rmtree(d)
    os.makedirs(d)

for class_name in CLASS_NAMES:
    class_path = os.path.join(ORIGINAL_TEST_DIR, class_name)
    all_files = [os.path.join(class_path, f) for f in os.listdir(class_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
    v_files, ts_files = train_test_split(all_files, test_size=0.5, random_state=SEED)
    for folder, files in [(VAL_DIR, v_files), (FINAL_TEST_DIR, ts_files)]:
        dest = os.path.join(folder, class_name); os.makedirs(dest, exist_ok=True)
        for f in files: shutil.copy(f, os.path.join(dest, os.path.basename(f)))
print("‚úÖ Phase 4 Complete: Data splitting finished.")

# --- PHASE 5: COMPREHENSIVE OPTUNA SEARCH ---
def optuna_objective(trial):
    lr = trial.suggest_float('lr', 1e-5, 1e-3, log=True)
    bs = trial.suggest_categorical('bs', [16, 32])
    opt_name = trial.suggest_categorical('optimizer', ['Adam', 'Adamax'])
    dropout = trial.suggest_float('dropout', 0.2, 0.4)
    base_f = trial.suggest_categorical('base_filters', C3K2_FILTERS_SEARCH)
    stages = trial.suggest_categorical('stages', BACKBONE_DEPTH_SEARCH)
    
    t, v, _ = get_loaders(bs)
    model = build_yolov11_classifier(IMAGE_SIZE + (3,), NUM_CLASSES, base_f, stages, dropout)
    
    optimizer = Adam(learning_rate=lr) if opt_name == 'Adam' else Adamax(learning_rate=lr)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    
    history = model.fit(t, epochs=EPOCHS_PER_TRIAL, validation_data=v, verbose=0, 
                        callbacks=[TFKerasPruningCallback(trial, 'val_accuracy')])
    return np.max(history.history['val_accuracy'])

print("üß† Starting Optuna Hyperparameter Search...")
study = optuna.create_study(direction='maximize')
study.optimize(optuna_objective, n_trials=N_TRIALS)
best_p = study.best_params
print(f"‚úÖ Best Trial Params: {best_p}")

# --- PHASE 6: FREEZE PERCENT SEARCH ---
print("‚ùÑÔ∏è Searching for Optimal Freeze Percentage...")
def optuna_freeze_objective(trial):
    freeze_p = trial.suggest_categorical('freeze_percent', FREEZE_PERCENT_SEARCH)
    t, v, _ = get_loaders(best_p['bs'])
    model = build_yolov11_classifier(IMAGE_SIZE + (3,), NUM_CLASSES, best_p['base_filters'], best_p['stages'], best_p['dropout'])
    
    num_to_freeze = int(len(model.layers) * (freeze_p / 100.0))
    for layer in model.layers[:num_to_freeze]: layer.trainable = False
    
    model.compile(optimizer=Adam(learning_rate=best_p['lr'] * 0.1), loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(t, epochs=5, validation_data=v, verbose=0)
    return np.max(history.history['val_accuracy'])

study_f = optuna.create_study(direction='maximize')
study_f.optimize(optuna_freeze_objective, n_trials=N_TRIALS_FREEZE)
best_freeze = study_f.best_params['freeze_percent']
print(f"‚úÖ Optimal Freeze: {best_freeze}%")

# --- PHASE 7: FINAL TWO-PHASE TRAINING ---
print("üöÄ Phase 7: Final Training Starting...")
train_gen, val_gen, test_gen = get_loaders(best_p['bs'])
yolo11_model = build_yolov11_classifier(IMAGE_SIZE + (3,), NUM_CLASSES, best_p['base_filters'], best_p['stages'], best_p['dropout'])

# Stage 1: Full Architecture Initial Training
print("üîπ Stage 1: Initial Training...")
opt = Adam(learning_rate=best_p['lr']) if best_p['optimizer'] == 'Adam' else Adamax(learning_rate=best_p['lr'])
yolo11_model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
h1 = yolo11_model.fit(train_gen, epochs=INITIAL_EPOCHS, validation_data=val_gen)

# Stage 2: Fine-Tuning with Freezing
print(f"üîπ Stage 2: Fine-Tuning (Freezing {best_freeze}% Layers)...")
num_to_freeze = int(len(yolo11_model.layers) * (best_freeze / 100.0))
for layer in yolo11_model.layers[:num_to_freeze]: layer.trainable = False
yolo11_model.compile(optimizer=Adam(learning_rate=best_p['lr'] * 0.01), loss='categorical_crossentropy', metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
h2 = yolo11_model.fit(train_gen, epochs=INITIAL_EPOCHS + FINE_TUNE_EPOCHS, initial_epoch=h1.epoch[-1]+1, validation_data=val_gen)
print("‚úÖ Phase 7 Complete: Training finished.")

# --- PHASE 8: EVALUATION ---
print("üìä Phase 8: Generating Final Performance Metrics...")
y_pred_probs = yolo11_model.predict(test_gen)
y_pred = np.argmax(y_pred_probs, axis=1)

print("\n--- CLASSIFICATION REPORT ---")
print(classification_report(test_gen.classes, y_pred, target_names=CLASS_NAMES, digits=4))

# Confusion Matrix
plt.figure(figsize=(8,6))
sns.heatmap(confusion_matrix(test_gen.classes, y_pred), annot=True, fmt='d', cmap='Blues', xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.title('YOLOv11 Final Confusion Matrix')
plt.ylabel('Actual Label'); plt.xlabel('Predicted Label'); plt.show()

# ROC-AUC Curve
plt.figure(figsize=(8,6))
for i, name in enumerate(CLASS_NAMES):
    fpr, tpr, _ = roc_curve(test_gen.classes == i, y_pred_probs[:, i])
    plt.plot(fpr, tpr, label=f'{name} (AUC = {auc(fpr, tpr):.4f})')
plt.plot([0, 1], [0, 1], 'k--'); plt.title('ROC Curves'); plt.legend(); plt.show()
```
