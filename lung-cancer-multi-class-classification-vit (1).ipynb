```python
!pip uninstall protobuf -y 
!pip install protobuf==3.20.3 
!pip install optuna-integration[tfkeras]
!pip install opencv-python --no-deps

import os
import shutil
import numpy as np
import tensorflow as tf
from tensorflow import keras 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import cycle
from tensorflow.keras.preprocessing.image import ImageDataGenerator # Still needed for flow_from_directory
from tensorflow.keras import layers, Model
from tensorflow.keras.optimizers import Adam, Adamax, RMSprop 
from optuna.integration.tfkeras import TFKerasPruningCallback
import optuna
import time 

# --- I. CONFIGURATION (ViT-Small Parameters) ---

KAGGLE_INPUT_PATH = '/kaggle/input/lung-cancer-dataset-with-augmentation'
DATASET_ROOT = os.path.join(KAGGLE_INPUT_PATH, 'Lung Cancer Dataset')
TRAIN_DIR = os.path.join(DATASET_ROOT, 'Train')
ORIGINAL_TEST_DIR = os.path.join(DATASET_ROOT, 'Test') 

WORKING_DIR = '/kaggle/working'
VAL_DIR = os.path.join(WORKING_DIR, 'Val_50_percent')
FINAL_TEST_DIR = os.path.join(WORKING_DIR, 'Final_Test_50_percent')

# Hyperparameters and settings 
IMAGE_SIZE = (224, 224) 
NUM_CLASSES = 3
SEED = 42
EPOCHS_PER_TRIAL = 10    
INITIAL_EPOCHS = 25    
FINE_TUNE_EPOCHS = 15 
N_TRIALS = 15          
N_TRIALS_FREEZE = 8    
FINAL_WEIGHT_DECAY = 0.0 

# ViT Specific Parameters
PATCH_SIZE = 16          
NUM_PATCHES = (IMAGE_SIZE[0] // PATCH_SIZE) ** 2  # 196
PROJECTION_DIM_DEFAULT = 384 
TRANSFORMER_LAYERS_DEFAULT = 8 
NUM_HEADS = 6          
FF_DIM = 1536 

# For Optuna Search Space 
PROJECTION_DIM_SEARCH = [256, 384, 512] 
TRANSFORMER_LAYERS_SEARCH = [6, 8, 10] 
FREEZE_PERCENT_SEARCH = [10, 20, 30, 40, 50, 70, 80, 90] 

np.random.seed(SEED)
tf.random.set_seed(SEED)
CLASS_NAMES = ['Bengin cases', 'Malignant cases', 'Normal cases']

print(f"Configuration complete. ViT-Small-like Parameters: Image Size={IMAGE_SIZE}, Patch Size={PATCH_SIZE}, Layers={TRANSFORMER_LAYERS_DEFAULT}")

# --- II. ViT ARCHITECTURE COMPONENTS ---

class Patches(layers.Layer):
    def __init__(self, patch_size, **kwargs):
        super().__init__(**kwargs)
        self.patch_size = patch_size

    def call(self, images):
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1, 1, 1, 1],
            padding='VALID',
        )
        patch_dims = patches.shape[-1]
        patches = layers.Reshape((-1, patch_dims))(patches)
        return patches

class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim, **kwargs):
        super().__init__(**kwargs)
        self.num_patches = num_patches
        self.projection = layers.Dense(units=projection_dim, name="projection")
        
        self.position_embedding = layers.Embedding(
            input_dim=num_patches + 1,
            output_dim=projection_dim,
            name="positional_embedding"
        )
        
        initializer = keras.initializers.RandomNormal(stddev=0.02)
        self.cls_token = tf.Variable(
            initial_value=initializer(shape=(1, 1, projection_dim), dtype="float32"),
            trainable=True,
            name="cls_token"
        )

    def call(self, patches):
        batch_size = tf.shape(patches)[0]
        patches_embed = self.projection(patches)
        
        cls_token_tiled = tf.tile(self.cls_token, [batch_size, 1, 1])
        encoded_patches = tf.concat([cls_token_tiled, patches_embed], axis=1)
        
        positions = tf.range(start=0, limit=self.num_patches + 1, delta=1)
        position_embeddings = self.position_embedding(positions)
        encoded_patches = encoded_patches + position_embeddings
        
        return encoded_patches

class TransformerBlock(layers.Layer):
    def __init__(self, projection_dim, num_heads, ff_dim, rate=0.1, **kwargs): 
        super().__init__(**kwargs)
        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
        self.mha = layers.MultiHeadAttention(num_heads=num_heads, key_dim=projection_dim, dropout=rate)
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)
        
        self.mlp = keras.Sequential([
            layers.Dense(ff_dim, activation=tf.nn.gelu), 
            layers.Dropout(rate), 
            layers.Dense(projection_dim),
            layers.Dropout(rate) 
        ])
        self.ff_dim = ff_dim

    def call(self, inputs, training=False):
        x = self.layernorm1(inputs)
        x = self.mha(query=x, value=x, key=x, attention_mask=None, training=training)
        x = layers.Add()([inputs, x])
        
        y = self.layernorm2(x)
        y = self.mlp(y, training=training)
        return layers.Add()([x, y])

print("ViT Architecture Components defined.")

# --- III. ViT Model Builder Function ---
def build_vit_model(input_shape, num_classes, patch_size, num_patches, projection_dim, transformer_layers, num_heads, ff_dim, dropout_rate): 
    """Builds the Vision Transformer (ViT) model."""
    
    inputs = keras.Input(shape=input_shape) 
    
    patches = Patches(patch_size)(inputs)
    x = PatchEncoder(num_patches, projection_dim, name="patch_encoder")(patches)

    for i in range(transformer_layers):
        x = TransformerBlock(projection_dim, num_heads, ff_dim, rate=dropout_rate, name=f'transformer_block_{i}')(x)
    
    representation = x[:, 0, :] 
    
    # MLP Head
    y = layers.LayerNormalization(epsilon=1e-6, name='layer_normalization')(representation) 
    
    y = layers.Dense(projection_dim * 2, activation=tf.nn.gelu, name='dense_1')(y)
    y = layers.Dropout(dropout_rate, name="dropout_mlp_head")(y) 
    
    output = layers.Dense(num_classes, activation="softmax", name='predictions')(y)

    model = keras.Model(inputs=inputs, outputs=output)
    return model

print("ViT model builder function 'build_vit_model' defined.")


# --- NEW FUNCTION: Layer Freezing for Keras ---

def freeze_vit_layers_tf(model, freeze_percent_int):
    """
    Keras/TensorFlow ViT ‡¶Æ‡¶°‡ßá‡¶≤‡ßá‡¶∞ ENCODER ‡¶¨‡ßç‡¶≤‡¶ï‡ßá‡¶∞ ‡¶™‡ßç‡¶∞‡¶•‡¶Æ ‡¶¶‡¶ø‡¶ï‡ßá‡¶∞ ‡¶≤‡ßá‡¶Ø‡¶º‡¶æ‡¶∞‡¶ó‡ßÅ‡¶≤‡ßã ‡¶´‡ßç‡¶∞‡¶ø‡¶ú ‡¶ï‡¶∞‡¶æ‡¶∞ ‡¶´‡¶æ‡¶Ç‡¶∂‡¶®‡•§
    """
    freeze_percent = freeze_percent_int / 100.0
    
    transformer_layers = [layer for layer in model.layers if layer.name.startswith('transformer_block_')]
    total_layers = len(transformer_layers)
    layers_to_freeze = int(total_layers * freeze_percent)

    print(f"Applying Freezing: Total ViT encoder layers: {total_layers}. Freezing the first {layers_to_freeze} layers ({freeze_percent_int}%).")

    for i, layer in enumerate(transformer_layers):
        if i < layers_to_freeze:
            layer.trainable = False
        else:
            layer.trainable = True

    for layer in model.layers:
        if layer.name.startswith('patch_encoder'):
            layer.trainable = False
            
        if layer.name in ['layer_normalization', 'dense_1', 'dropout_mlp_head', 'predictions']:
            layer.trainable = True

# --- IV. DATASET PREPARATION ---

# --- Cell 2: Dataset Splitting ---
print("\n## 2. üìÇ Splitting Original Test Data (50/50) into Val and Final Test Sets")

for d in [VAL_DIR, FINAL_TEST_DIR]:
    if os.path.exists(d):
        shutil.rmtree(d)
    os.makedirs(d)

for class_name in CLASS_NAMES:
    class_path = os.path.join(ORIGINAL_TEST_DIR, class_name)
    if not os.path.exists(class_path):
        print(f"Error: Directory not found: {class_path}")
        continue
    
    all_files = [os.path.join(class_path, f) for f in os.listdir(class_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
    
    val_files, test_files = train_test_split(
        all_files, 
        test_size=0.5, 
        random_state=SEED
    )
    
    os.makedirs(os.path.join(VAL_DIR, class_name), exist_ok=True)
    os.makedirs(os.path.join(FINAL_TEST_DIR, class_name), exist_ok=True)

    for f in val_files:
        shutil.copy(f, os.path.join(VAL_DIR, class_name, os.path.basename(f)))
    for f in test_files:
        shutil.copy(f, os.path.join(FINAL_TEST_DIR, class_name, os.path.basename(f)))

print(f"Split complete. New folders created: 'Val_50_percent' and 'Final_Test_50_percent'.")

# --- Cell 4: Data Split Verification and Data Loader Setup ---
print("\n## 4. üîç Data Split Verification")

# *** UPDATE: NO NORMALIZATION/PREPROCESSING IN DATAGENERATOR ***
train_datagen_raw = ImageDataGenerator() 
val_datagen_raw = ImageDataGenerator()
test_datagen_raw = ImageDataGenerator() 

train_gen_for_counts = train_datagen_raw.flow_from_directory(
    TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=32, class_mode='categorical', seed=SEED, shuffle=True
)
val_gen_for_counts = val_datagen_raw.flow_from_directory(
    VAL_DIR, target_size=IMAGE_SIZE, batch_size=32, class_mode='categorical', seed=SEED, shuffle=False
)
final_test_gen_for_counts = test_datagen_raw.flow_from_directory(
    FINAL_TEST_DIR, target_size=IMAGE_SIZE, batch_size=32, class_mode='categorical', seed=SEED, shuffle=False
)

print(f"  Total Images: Train Count: {train_gen_for_counts.n}, Validation Count: {val_gen_for_counts.n}, Test Count: {final_test_gen_for_counts.n}")

def get_class_counts(generator, set_name):
    idx_to_class = {v: k for k, v in generator.class_indices.items()}
    counts = np.bincount(generator.classes)
    
    print(f"üìä {set_name} Set:")
    for idx, count in enumerate(counts):
        if idx in idx_to_class:
            print(f"  - {idx_to_class[idx]}: {count} images")

print("\n--- Detailed Class Counts ---")
get_class_counts(train_gen_for_counts, "Train")
get_class_counts(val_gen_for_counts, "Validation")
get_class_counts(final_test_gen_for_counts, "Test")
print("Data loaders setup and counts verified.")


# --- V. OPTUNA HYPERPARAMETER SEARCH (PHASE 1 - AdamW included) ---

# --- Cell 5: Optuna Objective Function (Phase 1 Hyperparameter Search) ---
def optuna_objective(trial):
    
    lr = trial.suggest_float('learning_rate', 1e-5, 1e-3, log=True)
    batch_size = trial.suggest_categorical('batch_size', [16, 32, 64])
    optimizer_choices = ['Adam', 'Adamax', 'RMSprop', 'AdamW'] 
    dropout_rate = trial.suggest_float('dropout_rate', 0.2, 0.5)
    optimizer_name = trial.suggest_categorical('optimizer', optimizer_choices) 
    
    weight_decay = 0.0
    if optimizer_name == 'AdamW':
        weight_decay = trial.suggest_float('weight_decay', 1e-5, 1e-2, log=True)

    # *** UPDATE: NO NORMALIZATION/AUGMENTATION ***
    train_datagen = ImageDataGenerator()
    val_datagen = ImageDataGenerator() 
    
    try:
        train_generator = train_datagen.flow_from_directory(TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=batch_size, class_mode='categorical', seed=SEED, shuffle=True)
        val_generator = val_datagen.flow_from_directory(VAL_DIR, target_size=IMAGE_SIZE, batch_size=batch_size, class_mode='categorical', seed=SEED, shuffle=False)
    except Exception as e:
        print(f"Data loading error in trial: {e}")
        return 0.0

    # Model Build & Compile
    model = build_vit_model(IMAGE_SIZE + (3,), NUM_CLASSES, PATCH_SIZE, NUM_PATCHES, PROJECTION_DIM_DEFAULT, TRANSFORMER_LAYERS_DEFAULT, NUM_HEADS, FF_DIM, dropout_rate)
    
    if optimizer_name == 'Adam' or optimizer_name == 'AdamW':
        optimizer = Adam(learning_rate=lr, weight_decay=weight_decay) 
    elif optimizer_name == 'Adamax':
        optimizer = Adamax(learning_rate=lr)
    else: 
        optimizer = RMSprop(learning_rate=lr) 

    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    pruning_callback = TFKerasPruningCallback(trial, 'val_accuracy') 
    
    history = model.fit(
        train_generator,
        epochs=EPOCHS_PER_TRIAL, 
        validation_data=val_generator,
        callbacks=[pruning_callback], 
        verbose=0 
    )

    return np.max(history.history['val_accuracy'])

print("Optuna ViT objective function for Phase 1 defined (AdamW included).")


# --- Cell 6: Execute Optuna Hyperparameter Optimization (Phase 1) ---
print("\n## 6. üß† Starting Optuna Hyperparameter Optimization (Phase 1)")

study_vit = optuna.create_study(direction='maximize', study_name="initial_hparam_search_vit")
study_vit.optimize(optuna_objective, n_trials=N_TRIALS, show_progress_bar=True) 

best_params_vit = study_vit.best_params
best_val_accuracy_vit = study_vit.best_value

FINAL_WEIGHT_DECAY = best_params_vit.get('weight_decay', 0.0)

print("\n--- Optuna Study Results (Phase 1) ---")
print(f"Best Validation Accuracy: {best_val_accuracy_vit:.4f}")
print("Best Hyperparameters:")
print(best_params_vit)

FINAL_LR = best_params_vit['learning_rate']
FINAL_BATCH_SIZE = best_params_vit['batch_size']
FINAL_OPTIMIZER = best_params_vit['optimizer']
FINAL_DROPOUT = best_params_vit['dropout_rate']


# --- VI. OPTUNA ARCHITECTURAL TUNING (PHASE 2) ---

# --- Cell 7: Optuna Objective for ViT Architectural Tuning (Phase 2 Search) ---
def optuna_vit_arch_objective(trial):
    
    proj_dim = trial.suggest_categorical('projection_dim', PROJECTION_DIM_SEARCH)
    num_layers = trial.suggest_categorical('transformer_layers', TRANSFORMER_LAYERS_SEARCH)
    
    ff_dim_fixed = FF_DIM 

    lr = FINAL_LR
    batch_size = FINAL_BATCH_SIZE
    dropout_rate = FINAL_DROPOUT
    optimizer_name = FINAL_OPTIMIZER
    FINE_TUNE_LR = FINAL_LR * 0.1 

    # *** UPDATE: NO NORMALIZATION/AUGMENTATION ***
    train_datagen = ImageDataGenerator() 
    val_datagen = ImageDataGenerator() 
    
    train_generator = train_datagen.flow_from_directory(TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=batch_size, class_mode='categorical', seed=SEED, shuffle=True)
    val_generator = val_datagen.flow_from_directory(VAL_DIR, target_size=IMAGE_SIZE, batch_size=batch_size, class_mode='categorical', seed=SEED, shuffle=False)

    # 1. Model Build
    model = build_vit_model(IMAGE_SIZE + (3,), NUM_CLASSES, PATCH_SIZE, NUM_PATCHES, proj_dim, num_layers, NUM_HEADS, ff_dim_fixed, dropout_rate)
    
    if optimizer_name == 'Adam' or optimizer_name == 'AdamW':
        fine_tune_optimizer = Adam(learning_rate=FINE_TUNE_LR, weight_decay=FINAL_WEIGHT_DECAY)
    elif optimizer_name == 'Adamax':
        fine_tune_optimizer = Adamax(learning_rate=FINE_TUNE_LR)
    else: 
        fine_tune_optimizer = RMSprop(learning_rate=FINE_TUNE_LR)

    model.compile(optimizer=fine_tune_optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

    pruning_callback = TFKerasPruningCallback(trial, 'val_accuracy') 

    history = model.fit(
        train_generator,
        epochs=EPOCHS_PER_TRIAL, 
        validation_data=val_generator,
        callbacks=[pruning_callback], 
        verbose=0 
    )

    return np.max(history.history['val_accuracy'])

print("Optuna ViT objective function for Phase 2 (Architectural Tuning) defined.")


# --- Cell 8: Execute Optuna Optimization for ViT Architectural Tuning (Phase 2) ---
print("\n## 8. üß† Starting Optuna Optimization for ViT Architectural Tuning (Phase 2)")

study_arch_vit = optuna.create_study(direction='maximize', study_name="vit_arch_search")
study_arch_vit.optimize(optuna_vit_arch_objective, n_trials=N_TRIALS_FREEZE, show_progress_bar=True) 

best_arch_params_vit = study_arch_vit.best_params
best_arch_val_accuracy_vit = study_arch_vit.best_value

print("\n--- Optuna Study Results (Phase 2 - ViT Architecture) ---")
print(f"Best ViT Architecture Validation Accuracy: {best_arch_val_accuracy_vit:.4f}")
print("Best ViT Architecture Hyperparameter:")
print(best_arch_params_vit)

FINAL_PROJECTION_DIM = best_arch_params_vit['projection_dim']
FINAL_TRANSFORMER_LAYERS = best_arch_params_vit['transformer_layers']
FINAL_FF_DIM = FF_DIM # Fixed FF_DIM for final run


# --- VII. OPTUNA FREEZE PERCENT TUNING (PHASE 3 - NEW) ---

print("\n## 11. üß† Starting Optuna Optimization for ViT Layer Freezing (Phase 3)")

# --- Cell 11: Optuna Objective for Freeze Percent Tuning (Phase 3 Search) ---
def optuna_freeze_objective(trial):
    
    freeze_percent_int = trial.suggest_categorical('freeze_percent', FREEZE_PERCENT_SEARCH)
    
    # Use best parameters from Phase 1 and 2
    proj_dim = FINAL_PROJECTION_DIM
    num_layers = FINAL_TRANSFORMER_LAYERS
    dropout_rate = FINAL_DROPOUT
    optimizer_name = FINAL_OPTIMIZER
    
    batch_size = FINAL_BATCH_SIZE
    
    # *** UPDATE: NO NORMALIZATION/AUGMENTATION ***
    train_datagen = ImageDataGenerator() 
    val_datagen = ImageDataGenerator()
    
    train_generator = train_datagen.flow_from_directory(TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=batch_size, class_mode='categorical', seed=SEED, shuffle=True)
    val_generator = val_datagen.flow_from_directory(VAL_DIR, target_size=IMAGE_SIZE, batch_size=batch_size, class_mode='categorical', seed=SEED, shuffle=False)

    # 1. Model Build
    model = build_vit_model(IMAGE_SIZE + (3,), NUM_CLASSES, PATCH_SIZE, NUM_PATCHES, proj_dim, num_layers, NUM_HEADS, FF_DIM, dropout_rate)
    
    # Apply Layer Freezing
    freeze_vit_layers_tf(model, freeze_percent_int) 
    
    # 2. Compile Model with Fine-Tune LR (Low LR is essential for fine-tuning)
    FINE_TUNE_LR = FINAL_LR * 0.1
    
    if optimizer_name == 'Adam' or optimizer_name == 'AdamW':
        optimizer = Adam(learning_rate=FINE_TUNE_LR, weight_decay=FINAL_WEIGHT_DECAY)
    elif optimizer_name == 'Adamax':
        optimizer = Adamax(learning_rate=FINE_TUNE_LR)
    else:
        optimizer = RMSprop(learning_rate=FINE_TUNE_LR)
        
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

    pruning_callback = TFKerasPruningCallback(trial, 'val_accuracy') 

    history = model.fit(
        train_generator,
        epochs=EPOCHS_PER_TRIAL, 
        validation_data=val_generator,
        callbacks=[pruning_callback], 
        verbose=0 
    )

    return np.max(history.history['val_accuracy'])

# --- Cell 12: Execute Optuna Optimization for Layer Freezing (Phase 3) ---

study_freeze = optuna.create_study(direction='maximize', study_name="vit_freeze_search")
study_freeze.optimize(optuna_freeze_objective, n_trials=N_TRIALS_FREEZE, show_progress_bar=True) 

best_freeze_params = study_freeze.best_params
best_freeze_val_accuracy = study_freeze.best_value

print("\n--- Optuna Study Results (Phase 3 - Freeze Percent) ---")
print(f"Best Freeze Validation Accuracy: {best_freeze_val_accuracy:.4f}")
print("Best Freeze Percentage Hyperparameter:")
print(best_freeze_params)

FINAL_FREEZE_PERCENT = best_freeze_params['freeze_percent']


# --- VIII. FINAL MODEL TRAINING AND EVALUATION ---

# --- Cell 9: Final ViT Model Training (Three Phases) ---
print("\n## 9. üöÄ Final ViT Model Training (Phase 1: Initial & Phase 2: Fine-Tuning)")

# *** UPDATE: NO NORMALIZATION/AUGMENTATION ***
# --- 1. Final Data Generators ---
final_train_generator = ImageDataGenerator().flow_from_directory(
    TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=FINAL_BATCH_SIZE, class_mode='categorical', seed=SEED, shuffle=True
)
final_val_generator = ImageDataGenerator().flow_from_directory(
    VAL_DIR, target_size=IMAGE_SIZE, batch_size=FINAL_BATCH_SIZE, class_mode='categorical', seed=SEED, shuffle=False
)
final_test_generator = ImageDataGenerator().flow_from_directory(
    FINAL_TEST_DIR, target_size=IMAGE_SIZE, batch_size=FINAL_BATCH_SIZE, class_mode='categorical', seed=SEED, shuffle=False
)

print(f"Using best parameters: LR={FINAL_LR:.2e}, Batch={FINAL_BATCH_SIZE}, Optim={FINAL_OPTIMIZER}, Dropout={FINAL_DROPOUT:.2f}, WD={FINAL_WEIGHT_DECAY:.2e}")
print(f"ViT Arch: Proj Dim={FINAL_PROJECTION_DIM}, Layers={FINAL_TRANSFORMER_LAYERS}, Freeze={FINAL_FREEZE_PERCENT}%")

# --- 2. Build and Compile Final ViT Model (Phase 1 Setup) ---
final_vit_model = build_vit_model(
    input_shape=IMAGE_SIZE + (3,), 
    num_classes=NUM_CLASSES, 
    patch_size=PATCH_SIZE, 
    num_patches=NUM_PATCHES, 
    projection_dim=FINAL_PROJECTION_DIM, 
    transformer_layers=FINAL_TRANSFORMER_LAYERS, 
    num_heads=NUM_HEADS, 
    ff_dim=FINAL_FF_DIM, 
    dropout_rate=FINAL_DROPOUT
)

# Optimizer for Phase 1 (Initial LR)
if FINAL_OPTIMIZER == 'Adam' or FINAL_OPTIMIZER == 'AdamW':
    optimizer = Adam(learning_rate=FINAL_LR, weight_decay=FINAL_WEIGHT_DECAY)
elif FINAL_OPTIMIZER == 'Adamax':
    optimizer = Adamax(learning_rate=FINAL_LR)
else:
    optimizer = RMSprop(learning_rate=FINAL_LR)

final_vit_model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(name='precision'), tf.keras.metrics.Recall(name='recall')]
)

early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# --- 3. Phase 1: Initial Training (Encoder Layers are trainable here) ---
print(f"\nPhase 1: Initial training for up to {INITIAL_EPOCHS} epochs (Standard LR, All layers trainable)...")
history = final_vit_model.fit(
    final_train_generator,
    epochs=INITIAL_EPOCHS,
    validation_data=final_val_generator,
    callbacks=[early_stopping],
    verbose=1
)

# --- 4. Phase 2: Fine-Tuning (Lower LR and Layer Freezing) ---
FINE_TUNE_LR = FINAL_LR * 0.1 

# Apply Layer Freezing
freeze_vit_layers_tf(final_vit_model, FINAL_FREEZE_PERCENT)

# Optimizer for Phase 2 (Lower LR)
if FINAL_OPTIMIZER == 'Adam' or FINAL_OPTIMIZER == 'AdamW':
    fine_tune_optimizer = Adam(learning_rate=FINE_TUNE_LR, weight_decay=FINAL_WEIGHT_DECAY)
elif FINAL_OPTIMIZER == 'Adamax':
    fine_tune_optimizer = Adamax(learning_rate=FINE_TUNE_LR)
else:
    fine_tune_optimizer = RMSprop(learning_rate=FINE_TUNE_LR)

# Model Re-Compile (Changes in trainable layers require re-compilation)
final_vit_model.compile(
    optimizer=fine_tune_optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(name='precision'), tf.keras.metrics.Recall(name='recall')]
)

total_epochs = INITIAL_EPOCHS + FINE_TUNE_EPOCHS
initial_epoch_start = history.epoch[-1] + 1 

print(f"\nPhase 2: Fine-Tuning for up to {FINE_TUNE_EPOCHS} more epochs (Lower LR, {FINAL_FREEZE_PERCENT}% layers frozen)...")
history_fine_tune = final_vit_model.fit(
    final_train_generator,
    epochs=total_epochs,
    initial_epoch=initial_epoch_start,
    validation_data=final_val_generator,
    callbacks=[early_stopping],
    verbose=1
)


# --- Cell 10: Final Evaluation and Visualization ---
print("\n## 10. üìä Final Evaluation on the Reserved Final Test Set")

# 1. Final Model Evaluation
test_results = final_vit_model.evaluate(final_test_generator, verbose=1)

# Get predictions and true classes for detailed metrics
test_steps = final_test_generator.n // final_test_generator.batch_size + (final_test_generator.n % final_test_generator.batch_size != 0)
predictions = final_vit_model.predict(final_test_generator, steps=test_steps)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = final_test_generator.classes[:len(predicted_classes)]

# --- 2. Combine Training Histories ---
try:
    acc = history.history['accuracy'] + history_fine_tune.history['accuracy']
    val_acc = history.history['val_accuracy'] + history_fine_tune.history['val_accuracy']
    loss = history.history['loss'] + history_fine_tune.history['loss']
    val_loss = history.history['val_loss'] + history_fine_tune.history['val_loss']
    epochs_range = range(len(acc))
except NameError:
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs_range = range(len(acc))

print("Training histories combined for plotting.")

# --- 3. Accuracy and Loss Curves ---
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.title('ViT Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.title('ViT Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show() 

# --- 4. ROC Curve ---
print("\n--- 2. ROC Curve and AUC Results ---")

y_true_one_hot = tf.keras.utils.to_categorical(true_classes, num_classes=NUM_CLASSES)

plt.figure(figsize=(10, 8))
colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])
lw = 2
roc_auc = dict()

for i, color in zip(range(NUM_CLASSES), colors):
    fpr, tpr, _ = roc_curve(y_true_one_hot[:, i], predictions[:, i])
    roc_auc[i] = auc(fpr, tpr)
    
    plt.plot(
        fpr,
        tpr,
        color=color,
        lw=lw,
        label=f'ROC curve of class {CLASS_NAMES[i]} (AUC = {roc_auc[i]:.4f})'
    )

plt.plot([0, 1], [0, 1], 'k--', lw=lw)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ViT Receiver Operating Characteristic (ROC) Curve for Multi-Class')
plt.legend(loc="lower right")
plt.show() 

# --- 5. Classification Report & Confusion Matrix ---
print("\n--- 3. Final Classification Report ---")
print(classification_report(true_classes, predicted_classes, target_names=CLASS_NAMES, digits=4))

cm = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.title('ViT Confusion Matrix on Final Test Set')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show() 

print("\n--- ViT Implementation Complete ---")
```
