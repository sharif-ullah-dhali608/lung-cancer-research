```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.applications import EfficientNetB3
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from sklearn.preprocessing import label_binarize, StandardScaler

# --- ১. Model Preparation ---
base_model = EfficientNetB3(weights='imagenet', include_top=False, pooling='avg', input_shape=(300, 300, 3))

# --- ২. Feature Extraction ---
def extract_features(directory):
    features = []
    labels = []
    class_names = sorted(os.listdir(directory))
    
    for label_name in class_names:
        class_path = os.path.join(directory, label_name)
        if not os.path.isdir(class_path): continue
        
        print(f"Extracting features from {label_name}...")
        for img_name in os.listdir(class_path):
            try:
                img_path = os.path.join(class_path, img_name)
                img = load_img(img_path, target_size=(300, 300))
                img_array = img_to_array(img)
                img_array = np.expand_dims(img_array, axis=0)
                img_array = preprocess_input(img_array)
                
                feat = base_model.predict(img_array, verbose=0)
                features.append(feat.flatten())
                labels.append(label_name)
            except: continue
    return np.array(features), np.array(labels)

# Path setup
base_path = '/kaggle/input/lung-cancer-dataset-with-augmentation/Lung Cancer Dataset'

# Extract
train_feat, train_lab = extract_features(os.path.join(base_path, 'Train'))
test_feat, test_lab = extract_features(os.path.join(base_path, 'Test'))

# Combine
X_all = np.concatenate([train_feat, test_feat])
y_all = np.concatenate([train_lab, test_lab])

# --- ৩. Dataframe & Trimming ---
cols = [f'f{i+1}' for i in range(X_all.shape[1])]
df = pd.DataFrame(X_all, columns=cols)
df['label'] = y_all

# Preview
print("\n--- Dataframe Preview ---")
display(pd.concat([df.head(10), df.tail(10)]))

# Trimming to 3800
if len(df) > 3800:
    df = df.sample(n=3800, random_state=42).reset_index(drop=True)

print(f"Final Data Count: {len(df)}")

X = df.drop('label', axis=1).values
y = df['label'].values

# --- ৪. Splitting FIRST (No Leakage) ---
# Step A: 3800 -> Test (15%) = 570
X_temp, X_test_raw, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=y
)

# Step B: 3230 -> Valid (20%) = 646
X_train_raw, X_val_raw, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.20, random_state=42, stratify=y_temp
)

print(f"\nCounts -> Train: {X_train_raw.shape[0]}, Val: {X_val_raw.shape[0]}, Test: {X_test_raw.shape[0]}")

# --- ৫. PCA & Scaling ---
pca = PCA(n_components=512)

# Fit on Train ONLY
X_train_pca = pca.fit_transform(X_train_raw)
X_val_pca = pca.transform(X_val_raw)
X_test_pca = pca.transform(X_test_raw)

# Scaling (Crucial for SVM)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_pca)
X_val = scaler.transform(X_val_pca)
X_test = scaler.transform(X_test_pca)

print("PCA & Scaling Done.")

# --- ৬. SVM Implementation (Updated) ---
# class_weight='balanced' removed as requested
svm_model = SVC(kernel='rbf', C=10, gamma='scale', probability=True) 
svm_model.fit(X_train, y_train)

# Predictions
y_pred = svm_model.predict(X_test)
y_score = svm_model.predict_proba(X_test)

# --- ৭. Evaluation ---
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=np.unique(y_all), yticklabels=np.unique(y_all))
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curve
classes = np.unique(y_all)
y_test_bin = label_binarize(y_test, classes=classes)
n_classes = len(classes)

plt.figure(figsize=(10, 8))
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'ROC {classes[i]} (AUC = {roc_auc:0.4f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multi-class ROC Curve')
plt.legend(loc="lower right")
plt.show()

# Final Report
print("\nFinal Performance Report:")
print(f"Validation Accuracy: {svm_model.score(X_val, y_val)*100:.2f}%")
print(f"Test Accuracy: {accuracy_score(y_test, y_pred)*100:.2f}%")
print(classification_report(y_test, y_pred))
```
