```python
!pip install -U optuna optuna-integration[tfkeras]
import os
import shutil
import numpy as np
import pandas as pd
import optuna
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import cycle
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

# Keras Imports
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet import ResNet101, preprocess_input
from tensorflow.keras import layers, Model
from tensorflow.keras.optimizers import Adam, Adamax
from tensorflow.keras.regularizers import l2
from optuna.integration.tfkeras import TFKerasPruningCallback

# --- CELL 1: SETUP & CONFIG ---
KAGGLE_INPUT_PATH = '/kaggle/input/lung-cancer-dataset-with-augmentation'
DATASET_ROOT = os.path.join(KAGGLE_INPUT_PATH, 'Lung Cancer Dataset')
WORKING_DIR = '/kaggle/working'
VAL_DIR = os.path.join(WORKING_DIR, 'Val_50_percent')
FINAL_TEST_DIR = os.path.join(WORKING_DIR, 'Final_Test_50_percent')
TRAIN_DIR = os.path.join(DATASET_ROOT, 'Train')
ORIGINAL_TEST_DIR = os.path.join(DATASET_ROOT, 'Test')

IMAGE_SIZE = (224, 224) # Standard ResNet101 Input
NUM_CLASSES = 3
SEED = 42
EPOCHS_PER_TRIAL = 5   
INITIAL_EPOCHS = 35    
FINE_TUNE_EPOCHS = 30  
N_TRIALS = 15          
N_TRIALS_FREEZE = 10   

np.random.seed(SEED)
tf.random.set_seed(SEED)
CLASS_NAMES = ['Bengin cases', 'Malignant cases', 'Normal cases']

# --- CELL 2: DATA SPLITTING (50/50) ---
print("## 2. ğŸ“‚ Splitting Test Data into Val and Final Test...")
for d in [VAL_DIR, FINAL_TEST_DIR]:
    if os.path.exists(d): shutil.rmtree(d)
    os.makedirs(d)

for class_name in CLASS_NAMES:
    class_path = os.path.join(ORIGINAL_TEST_DIR, class_name)
    all_files = [os.path.join(class_path, f) for f in os.listdir(class_path) if f.endswith(('.png', '.jpg', '.jpeg'))]
    v_files, t_files = train_test_split(all_files, test_size=0.5, random_state=SEED)
    os.makedirs(os.path.join(VAL_DIR, class_name), exist_ok=True)
    os.makedirs(os.path.join(FINAL_TEST_DIR, class_name), exist_ok=True)
    for f in v_files: shutil.copy(f, os.path.join(VAL_DIR, class_name, os.path.basename(f)))
    for f in t_files: shutil.copy(f, os.path.join(FINAL_TEST_DIR, class_name, os.path.basename(f)))

# --- CELL 3: MODEL ARCHITECTURE ---
def build_resnet101_model(input_shape, num_classes, dropout_rate=0.4):
    base_model = ResNet101(weights='imagenet', include_top=False, input_shape=input_shape)
    base_model.trainable = False 
    x = base_model.output
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.BatchNormalization()(x)
    
    # Dense Units increased to capture subtle 'Bengin' features
    x = layers.Dense(1024, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = layers.Dropout(dropout_rate)(x)
    x = layers.Dense(512, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = layers.Dropout(dropout_rate/2)(x)
    
    output = layers.Dense(num_classes, activation='softmax', name='predictions')(x)
    return Model(inputs=base_model.input, outputs=output), base_model

# --- CELL 4: DATA LOADER SETUP (No Rescale) ---
# Using preprocess_input for architectural compatibility
datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

# --- CELL 5 & 6: OPTUNA PHASE 1 ---
def objective_p1(trial):
    lr = trial.suggest_float('learning_rate', 1e-5, 3e-4, log=True)
    bs = trial.suggest_categorical('batch_size', [16, 32])
    opt_n = trial.suggest_categorical('optimizer', ['Adam', 'Adamax'])
    dr = trial.suggest_float('dropout_rate', 0.2, 0.4)

    train_gen = datagen.flow_from_directory(TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=bs, class_mode='categorical', seed=SEED, shuffle=True)
    val_gen = datagen.flow_from_directory(VAL_DIR, target_size=IMAGE_SIZE, batch_size=bs, class_mode='categorical', seed=SEED, shuffle=False)

    model, _ = build_resnet101_model(IMAGE_SIZE + (3,), NUM_CLASSES, dr)
    opt = Adam(learning_rate=lr) if opt_n == 'Adam' else Adamax(learning_rate=lr)
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
    
    history = model.fit(train_gen, epochs=EPOCHS_PER_TRIAL, validation_data=val_gen, callbacks=[TFKerasPruningCallback(trial, 'val_accuracy')], verbose=0)
    return np.max(history.history['val_accuracy'])

study = optuna.create_study(direction='maximize')
study.optimize(objective_p1, n_trials=N_TRIALS)
best_p = study.best_params

# --- CELL 7 & 8: OPTUNA PHASE 2 (FREEZE COUNT) ---
def objective_p2(trial):
    fc = trial.suggest_int('freeze_count', 80, 200, step=30)
    train_gen = datagen.flow_from_directory(TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=best_p['batch_size'], class_mode='categorical', seed=SEED, shuffle=True)
    val_gen = datagen.flow_from_directory(VAL_DIR, target_size=IMAGE_SIZE, batch_size=best_p['batch_size'], class_mode='categorical', seed=SEED, shuffle=False)
    
    model, base_m = build_resnet101_model(IMAGE_SIZE + (3,), NUM_CLASSES, best_p['dropout_rate'])
    base_m.trainable = True 
    for layer in base_m.layers[:fc]: layer.trainable = False 
    
    model.compile(optimizer=Adam(learning_rate=best_p['learning_rate']*0.1), loss='categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(train_gen, epochs=5, validation_data=val_gen, verbose=0)
    return np.max(history.history['val_accuracy'])

f_study = optuna.create_study(direction='maximize')
f_study.optimize(objective_p2, n_trials=N_TRIALS_FREEZE)
best_p['freeze_count'] = f_study.best_params['freeze_count']

# --- CELL 9: FINAL TRAINING ---
train_gen = datagen.flow_from_directory(TRAIN_DIR, target_size=IMAGE_SIZE, batch_size=best_p['batch_size'], class_mode='categorical', seed=SEED, shuffle=True)
val_gen = datagen.flow_from_directory(VAL_DIR, target_size=IMAGE_SIZE, batch_size=best_p['batch_size'], class_mode='categorical', seed=SEED, shuffle=False)
test_gen = datagen.flow_from_directory(FINAL_TEST_DIR, target_size=IMAGE_SIZE, batch_size=best_p['batch_size'], class_mode='categorical', seed=SEED, shuffle=False)

model, base_m = build_resnet101_model(IMAGE_SIZE + (3,), NUM_CLASSES, best_p['dropout_rate'])
model.compile(optimizer=Adam(learning_rate=best_p['learning_rate']), loss='categorical_crossentropy', metrics=['accuracy'])
h1 = model.fit(train_gen, epochs=INITIAL_EPOCHS, validation_data=val_gen)

base_m.trainable = True
for layer in base_m.layers[:best_p['freeze_count']]: layer.trainable = False
model.compile(optimizer=Adam(learning_rate=best_p['learning_rate']*0.1), loss='categorical_crossentropy', metrics=['accuracy'])
h2 = model.fit(train_gen, epochs=INITIAL_EPOCHS+FINE_TUNE_EPOCHS, initial_epoch=h1.epoch[-1]+1, validation_data=val_gen)

# --- CELL 10: EVALUATION ---
preds = model.predict(test_gen)
y_pred = np.argmax(preds, axis=1)
y_true = test_gen.classes

print("\n--- Final Classification Report (ResNet101) ---")
print(classification_report(y_true, y_pred, target_names=CLASS_NAMES, digits=4))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6)); sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.title('Confusion Matrix: ResNet101'); plt.ylabel('True Label'); plt.xlabel('Predicted Label'); plt.show()

plt.figure(figsize=(10, 8))
colors = cycle(['aqua', 'darkorange', 'cornflowerblue'])
for i, color in zip(range(NUM_CLASSES), colors):
    fpr, tpr, _ = roc_curve(tf.keras.utils.to_categorical(y_true, NUM_CLASSES)[:, i], preds[:, i])
    plt.plot(fpr, tpr, color=color, lw=2, label=f'ROC {CLASS_NAMES[i]} - ResNet101 (AUC = {auc(fpr, tpr):.4f})')
plt.plot([0, 1], [0, 1], 'k--'); plt.title('Multi-Class ROC Curve: ResNet101'); plt.legend(loc="lower right"); plt.show()
```
